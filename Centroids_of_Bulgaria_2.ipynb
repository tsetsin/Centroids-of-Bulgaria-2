```python
from fiona import collection
from shapely.geometry import MultiPoint

with collection("provinces.geojson", "r") as input:
    for f in input:
        coordinates = f ['geometry']['coordinates']
        coordinates = coordinates[-1]
        points = MultiPoint(coordinates)
        point = points.centroid
        point = str(point)
        point = point[6:]
        point = point.replace(" ", "], [")
        point = point.replace("(", "pyplot.plot ([")
        point = point.replace(")", "], 'ro')")
        print (point)
```
